## 3. 데이터 타입

**데이터 type**

1. int(정수)
2. float(실수)
3. str(문자열) 
4. bool(참/거짓)


### 3-1. int(정수)


```python
a = 1
print(f"a의 타입: {type(a)}")
print(f"a의 값: {a}")

# 정수는 조건문에서 True로 취급됩니다
if 1:
    print('1은 참으로 취급')
else:
    print('1은 거짓부렁이')

if 0:
    print('0은 참으로 취급')
else:
    print('0은 거짓부렁이')

if 123:
    print('123은 참으로 취급')
else:
    print('123은 거짓부렁이')
```

### 3-2. float(실수)


```python
a = 3.14
print(f"a의 타입: {type(a)}")
print(f"a의 값: {a}")
```

### 3-3. str (문자열)


```python
# 작은따옴표와 큰따옴표 모두 사용 가능
word1 = '안녕하세요'
word2 = "안녕하세요"

print(f"word1의 타입: {type(word1)}")
print(f"word2의 타입: {type(word2)}")
print(f"word1의 값: {word1}")
print(f"word2의 값: {word2}")
```

### 3-4. bool (참/거짓)


```python
a = False
print(f"a의 값: {a}")
print(f"a의 타입: {type(a)}")

# 0과 False는 같습니다
print(f"0 == False: {0 == False}")

# True와 False
print(f"True: {True}")
print(f"False: {False}")
```

## 4. 데이터 타입 (집합)

**집합 형태의 데이터 타입**

1. list (순서 O, 집합)
2. tuple (순서 O, 읽기 전용 집합)
3. set (순서 X, 중복X 집합)
4. dict (key, value로 이루어진 사전형 집합)


### 4-1. list (순서가 있는 집합)

[] 형태로 표현합니다.


```python
# 빈 리스트 생성
mylist = []
print(f"빈 리스트: {mylist}")
print(f"리스트 타입: {type(mylist)}")

# 값이 있는 리스트
mylist = [1, 3, 2, 4, 5]
print(f"값이 있는 리스트: {mylist}")
```

#### 값을 추가하기


```python
mylist = []
print(f"시작: {mylist}")

# append()로 값 추가
mylist.append(1)
print(f"1 추가 후: {mylist}")

mylist.append(2)
mylist.append(3)
print(f"2, 3 추가 후: {mylist}")
```

#### 값을 제거하기


```python
print(f"현재 리스트: {mylist}")

# remove()로 값 제거 (첫 번째로 나타나는 값만 제거)
mylist.remove(1)
print(f"1 제거 후: {mylist}")

# 여러 개의 같은 값이 있을 때
mylist = [1, 2, 3, 1, 2, 3]
print(f"중복 값이 있는 리스트: {mylist}")

mylist.remove(1)  # 첫 번째 1만 제거
print(f"첫 번째 1 제거 후: {mylist}")

mylist.remove(1)  # 두 번째 1 제거
print(f"두 번째 1 제거 후: {mylist}")
```

#### 인덱싱(indexing) -> 색인

인덱스는 **0번 부터 시작** 합니다.


```python
mylist = [1, 2, 3, 4]
print(f"리스트: {mylist}")

# 인덱스로 접근
print(f"mylist[0]: {mylist[0]}")  # 첫 번째 요소
print(f"mylist[3]: {mylist[3]}")  # 네 번째 요소

# 인덱스로 값 변경
print(f"변경 전: {mylist}")
mylist[0] = 100
print(f"변경 후: {mylist}")

# 리스트 길이
print(f"리스트 길이: {len(mylist)}")
```

### 4-2. tuple (순서가 있는 집합, 읽기 전용)

() 로 표현합니다


```python
mytuple = (1, 2, 3, 4, 5)
print(f"튜플: {mytuple}")
print(f"튜플 타입: {type(mytuple)}")
print(f"튜플 길이: {len(mytuple)}")

# 튜플은 수정 불가능 (읽기 전용)
print("\n튜플은 다음 작업들이 불가능합니다:")
print("- append() 추가")
print("- remove() 제거") 
print("- 인덱스로 값 변경")

# mytuple.append(1)  # AttributeError 발생
# mytuple.remove(1)  # AttributeError 발생
# mytuple[0] = 1     # TypeError 발생
```

### 4-3. set (순서 X, 중복 X)


```python
# 빈 세트 생성
myset = set()
print(f"빈 세트: {myset}")
print(f"세트 타입: {type(myset)}")

# 값 추가
myset.add(1)
myset.add(2)
myset.add(3)
print(f"값 추가 후: {myset}")

# 중복 값 추가 (중복은 자동으로 제거됨)
myset.add(1)
myset.add(2)
myset.add(3)
print(f"중복 값 추가 후: {myset}")  # 변화 없음

# 새로운 값 추가
myset.add(4)
print(f"4 추가 후: {myset}")
```

### 4-4. dict (사전형 집합, key와 value 쌍)

{}로 표현합니다


```python
# 빈 딕셔너리 생성
mydict = dict()
print(f"빈 딕셔너리: {mydict}")
print(f"딕셔너리 타입: {type(mydict)}")

# 값 추가하기
mydict['apple'] = 123
print(f"apple 추가 후: {mydict}")

# 키로 값 접근
print(f"mydict['apple']: {mydict['apple']}")

# 숫자 키도 가능
mydict[0] = 2
print(f"숫자 키 0 추가 후: {mydict}")

# 다양한 타입의 키 사용
mydict[3.14] = 1
print(f"실수 키 3.14 추가 후: {mydict}")

# 값 변경
mydict['apple'] = 'byebye'
print(f"apple 값 변경 후: {mydict}")

# 딕셔너리 길이
print(f"딕셔너리 길이: {len(mydict)}")
```
